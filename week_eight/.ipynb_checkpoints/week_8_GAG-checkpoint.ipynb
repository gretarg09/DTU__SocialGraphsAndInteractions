# Path lengths

The first thing we want to take a look at is path lengths. NetworkX allows us to calculate the shortest path between any pair of articles. We begin by comparing the length of human and shortests paths.

### Exercise

** For each source/target pair in the list of human navigation paths, calculate the shortest path using NetworkX.**


We decided to use undirected graph


```python
# Read in the links.tsv file
file_path = '/Users/GretarAtli/Dropbox/Dtu/Social_graphs_and_interactions/wikispeedia_paths-and-graph/links.tsv'

G = nx.Graph()

with open(file_path, encoding="utf-8") as f:
    lines = f.readlines()
    lines = lines[12:]
    for line in lines:
            line = line.split()
            G.add_edge(line[0],line[1])
            
len(G.nodes())

all_paths_length = []

no_path = 0

for source in G.nodes():
    for target in G.nodes():
        if source != target:
            try:
                all_paths_length.append(len(nx.shortest_path(G,source=source,target=target)))
            except:
                no_path += 1
```

```python
# Used
counter = 100

# initialize the graph
G_human = nx.Graph()

sources = defaultdict(set)

# Read in the links.tsv file
file_path = '/Users/GretarAtli/Dropbox/Dtu/Social_graphs_and_interactions/wikispeedia_paths-and-graph/paths_finished.tsv'

with open(file_path, encoding="utf-8") as f:
    lines = f.readlines()
    lines = lines[16:] # Take out the lines that are not links
    for line in lines:
        line = line.split()[3] # Get the path from the line
        line = line.replace(";<","") # Back clicks are represented as "<", here we delete them from the path
        #print(line)
        #counter -= 1
        
        #if counter == 0:
        #    break
         
        line = line.split(";") # split the path on ; to get an array of websites
        source = line[0] # find the source website
        target = line[-1] # find the target website
        
        sources[source].add(target) # add the source as a key and append target to the corresponding target
        G_human.add_path(line) # Create the path in out network
        
        # For each line we add an edge to our graph
        #G.add_edge(line[0], line[1])            
        
```

```python
shortest_paths = []    

for source in sources.keys():
    for target in sources[source]:
        shortest_paths.append(len(nx.shortest_path(G,source=source,target=target)))
```

```python
plt.plot(list(Counter(shortest_paths).keys()), list(Counter(shortest_paths).values()))
plt.show()
```
