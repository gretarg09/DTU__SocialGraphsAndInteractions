** 2.17 ** Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
# let's start by importing the stopword
from nltk.corpus import stopwords
#stopwords.words('english')

def mostFrequent(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords] # Get non stopping words
    return FreqDist(content).most_common(50) # Use FreqDist to get 50 most common words
    
#FreqDist(text1).most_common(50)
mostFrequent(text1)
```

** 2.18 ** Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
from nltk import bigrams

def mostFrequentBigrams(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords] # Get non stopping words
    content_bigrams = bigrams(content)
    fdist = FreqDist(content_bigrams).most_common(50)
    return fdist

mostFrequentBigrams(text1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def processText(text, title):
    freqDist = [freq for word,freq in FreqDist(text).most_common()]
    
    plt.figure(figsize=(18,12))
    plt.plot(range(len(freqDist)),freqDist)
    plt.title("Zipf's Law - " + title)
    plt.xlabel("Word ranking")
    plt.ylabel("Frequency")
    
    plt.figure(figsize=(18,12))
    plt.figure(figsize=(18,12))
    plt.plot(range(len(freqDist)),freqDist)
    plt.xscale('log')
    plt.yscale('log')
    plt.title("Zipf's Law log log - " + title)
    plt.xlabel("Word ranking")
    plt.ylabel("Frequency")
    
    plt.show()

processText(text1, "text 1")

# Generate random text

randomstring = ""

for i in range(1000000):
    randomstring = randomstring + random.choice("abcdefg ")
    
processText(randomstring, "Randomly generated text")
```
