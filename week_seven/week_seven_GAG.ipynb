# Words that characterize the branches

In the exercise we'll use a very simple weighting scheme called TF-IDF to find the important words within each branch of philosophy (according to wikipedia). 

**First**, check out the wikipedia page for [TF-IDF](https://en.wikipedia.org/wiki/Tf%E2%80%93idf). Explain in your own words the point of TF-IDF.
    - What does TF stand for?
    - What does IDF stand for?
    
**Point of tf-idf** is to evaluate numerically how important a word is to a document in a collection or corpus. The ft-df value increases proportionally to the number of times a word appears in the document, but is often offset by the frequency of the word in the corpus, which helps to adjust for the fact that some words appear more frequently in general. 

tf-idf is the product of two statistics, term frequency and inverse document frequency. Note that there are various ways for determining the exect values of both statistics.

**Term frequency** tf(t,d), the simplest choice is to use the raw count of a term in a document, i.e. the number of times that term t occurs in document d. There are some other possibilities to evaluate the term frequency which can be found in the wikipedia page. 

** Inverse document frequency** is a measure of how much information the word provides, that is, whether the term is common or rare accross all documents. It is the logarithmically scaled inverse fraction of the documents that contain the word, obtained by dividing the total number of documents by number of documents containing the term, and then taking the logarithm of that quotient. See [wikipage](https://en.wikipedia.org/wiki/Tf%E2%80%93idf) for detail about equaltion
   

**Since we want to find out which words are important for each branch, so we're going to create six large documents, one per branch of philosophy. Tokenize the pages, and combine the tokens into one long list per branch. Remember the bullets below for success**

**1.** We start by getting the philosophers name and maintain the information about each philosopher in a dictionary

```python
import urllib2
import re

philosophers = {
    "aestheticians":{
        "title":"title=List_of_aestheticians",
        "names":"",
        "tokens":[]
    },
    "epistemologists":{
        "title":"title=List_of_epistemologists",
        "names":"",
        "tokens":[]
    },
    "ethicists":{
        "title":"title=List_of_ethicists",
        "names":"",
        "tokens":[]
    },
    "logicians":{
        "title":"title=List_of_logicians",
        "names":"",
        "tokens":[]
    },
    "metaphysicians":{
        "title":"title=List_of_metaphysicians",
        "names":"",
        "tokens":[]
    },
    "social and political philosophers":{
        "title":"title=Index_of_sociopolitical_thinkers",
        "names":"",
        "tokens":[]
    }
}

#'https://en.wikipedia.org/w/index.php?title=List_of_aestheticians&action=edit'

# set the parameters (explained in detail here https://www.mediawiki.org/wiki/API:Tutorial)
baseurl = "https://en.wikipedia.org/w/index.php?"
title = ""
action = "action=edit"

for i in philosophers:
    # construct the query
    query = "{}{}&{}".format(
    baseurl,
    philosophers[i]["title"],
    action)
    # use urllib and regex to get the list of philosophers
    wikiresponse = urllib2.urlopen(query)
    wikisource = wikiresponse.read()
    philosophers[i]["names"] = re.findall(r'\*.*?\[\[(.*?)[\]\|]', wikisource)

# Because the last 4 elements of the ethicists list and social and political philosophers are not philosophers
# they are left out
philosophers["ethicists"]["names"] = philosophers["ethicists"]["names"][:-4]
philosophers["social and political philosophers"]["names"] = philosophers["social and political philosophers"]["names"][:-4]

# Get all philosophers into one list
all_philo = []
for i in philosophers:
    all_philo = all_philo + philosophers[i]["names"]
all_philo = set(all_philo) # erase all duplicates
```

**2.** Next we go through all of the philosophers in each branch and we create a large document for each brand. Therefore we end up with six large documents that contain all the text of each brands philosopher´s wikipedia pages. The workflow is as follows:
    - Tokenize the pages
    - combine the tokens into a long list per branch

```python
import json
import io
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords




# set the parameters for the query (explained in detail here https://www.mediawiki.org/wiki/API:Tutorial)
baseurl = "https://en.wikipedia.org/w/api.php?"
actions = "action=query"
title = "titles="
#content = "prop=revisions&rvprop=content"
content = "prop=extracts&exlimit=max&explaintext&rvprop=content"
dataformat = "format=json"

# iterate through each philosophers brand
# The list comprehension goes through the list of philosophers and replaces " " for "_"
for brand in philosophers:
    for philosopher in philosophers[brand]["names"]:
        query = "{}{}&{}{}&{}&{}".format(baseurl,actions,title,philosopher.replace(" ","_"),content,dataformat)
        print query
        wikiresponse = urllib2.urlopen(query)
        wikisource = wikiresponse.read()
        
        # get the text from wikipage and change upper case symbols to lower case
        text =  unicode(json.dumps(wikisource)).lower()
        
        # We start our data cleaning by removing all \n from the text. We simply replace \n with " "
        text = re.sub(r"\n+"," ",text)
        text = re.sub(r"\\n+"," ",text)
        text = re.sub(r"\\u+"," ",text)
        
        tokenizer = RegexpTokenizer(r'\w+')  # We choose to use regularexpression tokenizer
        tokens = tokenizer.tokenize(text) # Get all tokens with the tokenizer
        # filter out the stop words and digits
        filtered_words = [w for w in tokens if not w in stopwords.words('english') if not w.isdigit()] 
        
        # TODO taka út nöfn á gaurum
        
        # Finally we use stemming 
        porter  = nltk.PorterStemmer()
        #lancaster = nltk.LancasterStemmer()

        filtered_words = [porter.stem(word) for word in filtered_words]
        #filtered_words = [lancaster.stem(word) for word in filtered_words]
        
        # Get unique set of the words
        filtered_words =  set(filtered_words)
        
        # Add the tokens from this specific wikipage to the dictionary
        philosophers[brand]["tokens"].append(filtered_words)
        
```
